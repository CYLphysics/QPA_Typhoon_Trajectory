```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.arange(10, 100, 10)
plt.figure(figsize=(8,4), dpi=300)

plt.subplot(1,2,1)

plt.title("High efficient linear models ($c=1)$")
plt.plot(x, x*0.092, label = "Orion series 100-200 qubits (QPA)", color = plt.cm.magma(0.55), marker = "o", lw = 1, markerfacecolor=plt.cm.magma(0.65))
plt.plot(x, x*0.0816, label = "Post-Orion series 400-1000 qubits (QPA)",  color = plt.cm.magma(0.12), marker = "^", lw = 1, markerfacecolor=plt.cm.magma(0.25))
plt.plot(x, x*0.0872, label = "Basic GPU server (Classical full model)", marker = "s", lw = 1, alpha = 0.7)
plt.plot(x, x*1.7130, label = "Joliot-Curie Rome (Classical full model)", color = plt.cm.magma(0.70),marker = "^", lw = 1, markerfacecolor=plt.cm.magma(0.90))
plt.xlabel("model size factor (runtime)")
plt.ylabel("tCO2 eq")
plt.yscale('log')
plt.legend(loc=4, fontsize = 7)

plt.subplot(1,2,2)

plt.title("Attention-based models ($c=2$)")
plt.plot(x, x*0.0022, label = "(QPA) Orion series 100-200 qubits", color = plt.cm.magma(0.55), marker = "o", lw = 1, markerfacecolor=plt.cm.magma(0.65))
plt.plot(x, x*0.0020, label = "(QPA) Post-Orion series 400-1000 qubits",  color = plt.cm.magma(0.12), marker = "^", lw = 1, markerfacecolor=plt.cm.magma(0.25))
plt.plot(x, x*0.0872, label = "Basic GPU server", marker = "s", lw = 1, alpha = 0.7)
plt.plot(x, x*1.7130, label = "Joliot-Curie Rome", color = plt.cm.magma(0.70),marker = "^", lw = 1, markerfacecolor=plt.cm.magma(0.90))
plt.xlabel("model size factor (runtime)")
# plt.ylabel("tCO2 eq")
plt.yscale('log')
plt.tight_layout()
plt.savefig("../results/plots/tco2_emission.png", bbox_inches="tight")  # Save as a PDF

# plt.legend()
```
